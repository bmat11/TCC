# Importando bibliotecas

```python
import torch
import json
from ultralytics import YOLO
from datetime import datetime
```

# Modelo Yolo v8n



## Verificando torch se reconhece CUDA

```python
print(torch.__version__)
print("CUDA disponível?", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "Nenhuma")
```

## Treinamento

```python
# Caminho para o arquivo YAML do dataset
yaml_path = r"C:\Users\evosystem03.ti\Downloads\TCC\ptld_yolo\ptld.yaml"

# Carregar o modelo YOLOv8n
model = YOLO("yolov8n.pt")

# Treinamento
results_train = model.train(
    data=yaml_path,
    epochs=150,       # número de épocas
    imgsz=384,        # tamanho das imagens
    batch=16,         # tamanho do batch
    device=0          # usar GPU (0 = primeira GPU)
)

print("✅ Treinamento concluído!")
```

## Validação

```python
# Validação do modelo treinado
metrics_val = model.val(
    data=yaml_path,
    imgsz=384,
    batch=16,
    device=0
)

# Exibir resultados de validação
print("📊 Resultados da Validação:")
print(metrics_val.results_dict)
```

## Teste

```python
# Carregar o melhor modelo treinado
model = YOLO(r"C:\Users\evosystem03.ti\Downloads\TCC\yolov8n\runs\detect\train\weights\best.pt")

# Avaliar o modelo no conjunto de teste
metrics_test = model.val(
    data=yaml_path,
    split="test",   # garante que usará o conjunto "test" do data.yaml
    imgsz=384,
    batch=16,
    device=0
)

# Exibir resultados do teste
print("📈 Resultados do Teste:")
print(metrics_test.results_dict)
```

## Salvando resultados

```python
# Dicionário com os resultados
results = {
    "data_execucao": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "val": metrics_val.results_dict,
    "test": metrics_test.results_dict
}

# Caminhos dos arquivos de saída
saida_txt = "resultados_yolov8n.txt"
saida_json = "resultados_yolov8n.json"

# --- Salvar em TXT ---
with open(saida_txt, "w", encoding="utf-8") as f:
    f.write("=== RESULTADOS YOLOv8n ===\n\n")
    f.write(f"Data de execução: {results['data_execucao']}\n\n")
    f.write("=== Validação ===\n")
    for k, v in results["val"].items():
        f.write(f"{k}: {v}\n")
    f.write("\n=== Teste ===\n")
    for k, v in results["test"].items():
        f.write(f"{k}: {v}\n")

# --- Salvar em JSON ---
with open(saida_json, "w", encoding="utf-8") as f:
    json.dump(results, f, indent=4, ensure_ascii=False)

print(f"✅ Resultados salvos em:\n- {saida_txt}\n- {saida_json}")
```
